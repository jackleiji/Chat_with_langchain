```python
# from langchain_community.vectorstores import Chroma
from langchain_chroma  import Chroma
from langchain.document_loaders import PyMuPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from langchain.llms import OpenAI
from langchain.llms import HuggingFacePipeline
from langchain_community.embeddings.sentence_transformer import SentenceTransformerEmbeddings
```

```python
import os
import openai

openai.api_key  = os.environ['OPENAI_API_KEY']
embedding = SentenceTransformerEmbeddings(model_name="E:/CATL/2.项目与比赛/m3e-base")
# embedding = OpenAIEmbeddings()
```

```python
persist_directory = r'E:\CATL\2.项目与比赛\Chat_with_Datawhale_langchain\vector_db'
vectordb = Chroma(
    persist_directory=persist_directory,
    embedding_function=embedding
)
print(f"向量库中存储的数量：{vectordb._collection.count()}")
```

```python
question="llm_universe"
sim_docs = vectordb.similarity_search(question,k=10)
print(f"检索到的内容数：{len(sim_docs)}")
```

```python
for i, sim_doc in enumerate(sim_docs):
    print(f"检索到的第{i}个内容: \n{sim_doc.page_content}", end="\n--------------\n")
```

```python
import requests
import loguru
def check_connection():
    try:
        requests.get('https://api.github.com', timeout=5)
        return True
    except:
        return False

# 在主函数中使用
if not check_connection():
    print("无法连接到GitHub API，请检查网络连接")
    loguru.logger.error("无法连接到GitHub API，请检查网络连接")
    exit(1)
```
